# Task 3
* Handle missing values (identify, fill/drop)

```python
'''Displaying null values in dataframe(true -> Null, False -> Not Null)'''

df.isnull()
```

```python
'''Displaying count of total null-values column wise'''

df.isnull().sum()

```

* Dropping Missing Values

```python
'''Example of dropping null values row wise'''
df.dropna(axis=0) 
```

```python
'''Example of dropping null values column wise'''
df.dropna(axis=1)
```

```python
df_filled = df.fillna(0)
df_filled

```

# Task 4
* Perform basic operations (filter rows, select columns)

```python
df_filled[(df_filled['rating']>=3.0) & (df_filled['rating']<=5.0)]
```

```python
df_filled['final_price'] = pd.to_numeric(
    df_filled['final_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
    .str.strip(),
    errors='coerce'
)
df_filled
```

```python
df_filled[(df_filled['final_price'] > 500) & (df_filled['final_price'] < 1000)]
```

```python
df_filled[["title","product_description","rating","final_price","delivery_options","discount"]]
```

# Task 6
* Create a derived column (total_amount = price * quantity). 

```python
df_filled
```

```python
df_filled.info()
```

```python
import random
df_filled['quantity'] = [random.randint(1,10) for i in range(len(df_filled))]
df_filled['Total_amount'] = df_filled['final_price'] * df_filled['quantity']
df_filled
```

* Graphical Representation


```python
numeric_columns = [
    'final_price',
    'rating',
    'discount',
    'quantity',
    'Total_amount'
]

plt.figure(figsize=(15,10))

for i, column in enumerate(numeric_columns, 1):

    plt.subplot(3, 2, i)

    df_filled[column].hist(bins=20)

    plt.title(column)

plt.tight_layout()

plt.show()
```
